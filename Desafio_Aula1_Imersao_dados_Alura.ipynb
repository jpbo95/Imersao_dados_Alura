Podemos considerar que nesses casos outras drogas não seriam eficientes e foi utilizado a droga "cacb2b860" para ter tal informação, portanto, o tratamento com controle se mostrou mais eficaz.

##Desafio 07: Deixar os graficos bonitos. (Matplotlib.pyplot)
